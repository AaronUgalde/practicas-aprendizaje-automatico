```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
positivos = [(5, 7), (6, 6), (5, 5), (4, 5), (4, 6)]
negativos = [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2)]

promedio_positivos = (
    sum(p[0] for p in positivos) / len(positivos),
    sum(p[1] for p in positivos) / len(positivos)
)

promedio_negativos = (
    sum(n[0] for n in negativos) / len(negativos),
    sum(n[1] for n in negativos) / len(negativos)
)

print("promedio_positivos:", promedio_positivos)
print("promedio_negativos:", promedio_negativos)
```

```python
plt.figure()

px, py = zip(*positivos)
nx, ny = zip(*negativos)

plt.scatter(px, py)
plt.scatter(nx, ny)

plt.quiver(0, 0, promedio_positivos[0], promedio_positivos[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, promedio_negativos[0], promedio_negativos[1], angles='xy', scale_units='xy', scale=1)

plt.title("Puntos y vectores promedio")

plt.show()
```

```python
suma_vectores = (promedio_positivos[0] + promedio_negativos[0], promedio_positivos[1] + promedio_negativos[1])
resta_vectores = (promedio_positivos[0] - promedio_negativos[0], promedio_positivos[1] - promedio_negativos[1])

plt.figure()

px, py = zip(*positivos)
nx, ny = zip(*negativos)

plt.scatter(px, py)
plt.scatter(nx, ny)

plt.quiver(0, 0, promedio_positivos[0], promedio_positivos[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, promedio_negativos[0], promedio_negativos[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, suma_vectores[0], suma_vectores[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(promedio_negativos[0], promedio_negativos[1], resta_vectores[0], resta_vectores[1], angles='xy', scale_units='xy', scale=1, color='blue')

plt.xlim(0, 8)
plt.ylim(0, 8)

plt.title("Puntos y vectores promedio")

plt.show()
```

```python
def intersect(p1, p2, p3, p4):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    x4, y4 = p4
    
    den = (x1 - x2)*(y3 - y4) - (y1 - y2)*(x3 - x4)
    if den == 0:
        return None  # paralelas o coinciden
    
    px = ((x1*y2 - y1*x2)*(x3 - x4) - (x1 - x2)*(x3*y4 - y3*x4)) / den
    py = ((x1*y2 - y1*x2)*(y3 - y4) - (y1 - y2)*(x3*y4 - y3*x4)) / den
    
    return (px, py)

def perpendicular_line(point, vector):
    x0, y0 = point
    a, b = vector
    
    perp = (-b, a)
    
    if perp[0] != 0:
        m = perp[1] / perp[0]
        c = y0 - m * x0
        return m, c, perp
    else:
        return None, x0, perp

intersection = intersect((0, 0), suma_vectores, promedio_negativos, promedio_positivos)
m, c, perp = perpendicular_line(intersection, suma_vectores)
x = np.linspace(-10, 10, 100)
y = m*x + c

suma_vectores = (promedio_positivos[0] + promedio_negativos[0], promedio_positivos[1] + promedio_negativos[1])
resta_vectores = (promedio_positivos[0] - promedio_negativos[0], promedio_positivos[1] - promedio_negativos[1])

plt.figure()

plt.plot(x, y)

px, py = zip(*positivos)
nx, ny = zip(*negativos)

plt.scatter(px, py)
plt.scatter(nx, ny)
plt.scatter(intersection[0], intersection[1])

plt.quiver(0, 0, promedio_positivos[0], promedio_positivos[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, promedio_negativos[0], promedio_negativos[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, suma_vectores[0], suma_vectores[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(promedio_negativos[0], promedio_negativos[1], resta_vectores[0], resta_vectores[1], angles='xy', scale_units='xy', scale=1, color='blue')

plt.xlim(0, 8)
plt.ylim(0, 8)

plt.title("Puntos y vectores promedio")

plt.show()
```
