```python
import pandas as pd
import numpy as np
```

```python
entrenamiento = {'x': [17, 3, 19, 15, 2, 16, 10]
           , 'y': [10.5, 2.1, 11.7, 9.3, 1.5, 9.9, 6.3]}

prueba = {'x': [79, 43, 87],
          'y': [47.7, 26.1, 52.5]}

df_entrenamiento = pd.DataFrame(entrenamiento)
df_prueba = pd.DataFrame(prueba)

sum_y = df_entrenamiento['y'].sum()
sum_x = df_entrenamiento['x'].sum()
sum_xy = (df_entrenamiento['x'] * df_entrenamiento['y']).sum()
sum_x2 = (df_entrenamiento['x'] ** 2).sum()
n = len(df_entrenamiento)


b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b0 = (sum_y - b1 * sum_x) / n

print(f"b0: {b0}, b1: {b1}")
print("sum_x:", sum_x, " sum_y:", sum_y, " sum_xy:", sum_xy, " sum_x2:", sum_x2, " n:", n)

df_prueba['y_pred'] = b0 + b1 * df_prueba['x']
for(i, row) in df_prueba.iterrows():
    print(f"x: {row['x']}, y_real: {row['y']}, y_pred: {row['y_pred']}, error: {abs(row['y'] - row['y_pred'])}")
```
