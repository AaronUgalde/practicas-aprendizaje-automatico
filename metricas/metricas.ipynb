```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Leer CSV
df = pd.read_csv('metricas.csv')

# Obtener clases reales y predichas
y_true = df['Clases real']
y_pred = df['Clases predicha']

# Calcular matriz de confusi√≥n
matriz = confusion_matrix(y_true, y_pred, labels=['A','E','I','O'])

# Mostrar matriz de forma tabular con pandas
matriz_df = pd.DataFrame(matriz, index=['A','E','I','O'], columns=['A','E','I','O'])
print(matriz_df)

report = classification_report(y_true, y_pred, labels=['A','E','I','O'])
print(report)
```
