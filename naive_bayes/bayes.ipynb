```python
import pandas as pd
import numpy as np
from collections import defaultdict
from sklearn.model_selection import train_test_split
import math
```

```python
entrenamiento = {'x': ["just plain boring", "entirely predictable and lacks energy", "no surprises and very few laughs", "very powerful", "the most fun film of the summer"]
           , 'y': [0, 0, 0, 1, 1]}

prueba = {'x': ["predictable with no fun", "very fun"],
          'y': [47.7, 26.1]}
```

```python
dataset = pd.read_csv("Book1.csv")
entrenamiento, prueba = train_test_split(dataset, test_size=0.2, shuffle=False)
print(entrenamiento)
diferent_words = set()
diferent_words_class = defaultdict(dict)
number_words_class = defaultdict(int)
```

```python
for x, y in zip(entrenamiento['Words'], entrenamiento['Clase']):
    words = x.split(' ')
    for word in words:
        if word not in diferent_words_class[y]:
            diferent_words_class[y][word]  = 0;
        diferent_words_class[y][word] += 1
        number_words_class[y] += 1;
        diferent_words.add(word)
```

```python
prob_clase = dict()
for clase in set(entrenamiento['Clase']):
    prob_clase[clase] = (entrenamiento['Clase'] == clase).sum() / len(entrenamiento)
```

$\alpha$

```python
V = len(diferent_words)
alpha = 1
word_probability_class = defaultdict(dict)
for word in diferent_words:
    for clase in set(entrenamiento['Clase']):
        p_word_clase = 0;
        if word in diferent_words_class[clase]:
            p_word_clase = diferent_words_class[clase][word] 
        word_probability_class[word][clase] = (p_word_clase + alpha) / (number_words_class[clase] + V)
        print(word, clase, " : ", word_probability_class[word][clase],  " cuantas veces aparace: ",    p_word_clase, " palabras diferentes en la clase: ", len(diferent_words_class[clase]), " v: ", V)
```

```python
for x in prueba['Words']:
    for clase in set(entrenamiento['Clase']):
        prob = prob_clase[clase]
        words = x.split(' ')
        for word in words:
            #print(x, " ", clase, " ", prob, " word: ", word)
            probabilidad_palabra = 1;
            if word in word_probability_class:
                probabilidad_palabra = word_probability_class[word][clase]
            prob *= probabilidad_palabra
        print(x, " ", clase, " ", prob, " log: ", math.log(prob))
    
```
